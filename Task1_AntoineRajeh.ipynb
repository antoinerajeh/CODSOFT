```python
import tkinter as tk
from tkinter import messagebox
from tkinter import simpledialog  # Add this line to import simpledialog

class TodoList:
    def __init__(self, master):
        self.master = master
        self.master.title("To-Do List")

        self.tasks = []

        self.create_widgets()

    def create_widgets(self):
        # Entry field for task input
        self.task_entry = tk.Entry(self.master, width=50)
        self.task_entry.grid(row=0, column=0, padx=5, pady=5)

        # Button to add a task
        self.add_button = tk.Button(self.master, text="Add Task", command=self.add_task)
        self.add_button.grid(row=0, column=1, padx=5, pady=5)

        # Listbox to display tasks
        self.task_listbox = tk.Listbox(self.master, width=50)
        self.task_listbox.grid(row=1, column=0, columnspan=2, padx=5, pady=5)

        # Buttons for updating and deleting tasks
        self.update_button = tk.Button(self.master, text="Update Task", command=self.update_task)
        self.update_button.grid(row=2, column=0, padx=5, pady=5)

        self.delete_button = tk.Button(self.master, text="Delete Task", command=self.delete_task)
        self.delete_button.grid(row=2, column=1, padx=5, pady=5)

    def add_task(self):
        task = self.task_entry.get()
        if task:
            self.tasks.append(task)
            self.task_listbox.insert(tk.END, task)
            self.task_entry.delete(0, tk.END)
        else:
            messagebox.showerror("Error", "Please enter a task.")

    def update_task(self):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            selected_task = self.task_listbox.get(selected_index)
            updated_task = simpledialog.askstring("Update Task", f"Enter updated task for '{selected_task}':")
            if updated_task:
                self.tasks[selected_index[0]] = updated_task
                self.task_listbox.delete(selected_index)
                self.task_listbox.insert(selected_index, updated_task)
        else:
            messagebox.showerror("Error", "Please select a task to update.")

    def delete_task(self):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            confirmed = messagebox.askyesno("Confirm Deletion", "Are you sure you want to delete this task?")
            if confirmed:
                del self.tasks[selected_index[0]]
                self.task_listbox.delete(selected_index)
        else:
            messagebox.showerror("Error", "Please select a task to delete.")

def main():
    root = tk.Tk()
    app = TodoList(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
