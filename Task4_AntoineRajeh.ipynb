```python
import tkinter as tk
from tkinter import messagebox
import random

class RockPaperScissorsGame:
    def __init__(self, master):
        self.master = master
        self.master.title("Rock-Paper-Scissors Game")
        
        self.user_score = 0
        self.computer_score = 0
        
        self.user_choice_var = tk.StringVar()
        
        self.create_widgets()
    
    def get_computer_choice(self):
        return random.choice(['rock', 'paper', 'scissors'])
    
    def determine_winner(self, user_choice, computer_choice):
        if user_choice == computer_choice:
            return 'tie'
        elif (user_choice == 'rock' and computer_choice == 'scissors') or \
             (user_choice == 'scissors' and computer_choice == 'paper') or \
             (user_choice == 'paper' and computer_choice == 'rock'):
            return 'user'
        else:
            return 'computer'
    
    def play_round(self):
        user_choice = self.user_choice_var.get()
        computer_choice = self.get_computer_choice()
        winner = self.determine_winner(user_choice, computer_choice)
        
        result_text = f"Computer's choice: {computer_choice}\n"
        if winner == 'tie':
            result_text += "It's a tie!"
        elif winner == 'user':
            result_text += "You win!"
            self.user_score += 1
        else:
            result_text += "Computer wins!"
            self.computer_score += 1
        
        result_text += f"\nUser Score: {self.user_score}\nComputer Score: {self.computer_score}"
        messagebox.showinfo("Result", result_text)
    
    def create_widgets(self):
        choices_frame = tk.Frame(self.master)
        choices_frame.pack(pady=10)
        
        tk.Label(choices_frame, text="Choose:").grid(row=0, column=0)
        rock_button = tk.Button(choices_frame, text="Rock", command=lambda: self.user_choice_var.set("rock"))
        rock_button.grid(row=0, column=1, padx=5)
        paper_button = tk.Button(choices_frame, text="Paper", command=lambda: self.user_choice_var.set("paper"))
        paper_button.grid(row=0, column=2, padx=5)
        scissors_button = tk.Button(choices_frame, text="Scissors", command=lambda: self.user_choice_var.set("scissors"))
        scissors_button.grid(row=0, column=3, padx=5)
        
        play_button = tk.Button(self.master, text="Play", command=self.play_round)
        play_button.pack(pady=5)
        
        play_again_button = tk.Button(self.master, text="Play Again", command=self.reset)
        play_again_button.pack(pady=5)
    
    def reset(self):
        self.user_score = 0
        self.computer_score = 0
        self.user_choice_var.set("")

def main():
    root = tk.Tk()
    game = RockPaperScissorsGame(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
